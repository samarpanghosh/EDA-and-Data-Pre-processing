# We can apply a custom function that we created on a single or multiple columns to create new columns
## Sometimes you have multiple conditions and you want to apply a function to multiple columns at the same time. 
## It can be done by using a custom made function, and applying this function to your dataframe.

```python
# It might be useful to know the split between males,females,and children

# We'll treat anyone as under 16 as a child, and then use the apply technique with a function to create a new column

def male_female_child(passenger):# Here passenger is a variable(i.e. passenger takes value of every entry in the passed columns)
    age,sex=passenger
# Above line means passenger iterates through every entry in the columns age and sex
    if age<18:
        return 'child'
    else:
        return sex      # Compare age otherwise leave the sex
    
```

## Now creating a column- 'person' and the entries of this new column would be derived by applying this custom made function to the age,sex columns 
## Axis should be set to 1 for columns and not index

```python
train_df['Person'] = train_df[['Age','Sex']].apply(male_female_child,axis=1)
train_df.head(20)
```

```python
# Now creating the catplot again but this time based on the newly created column 
sns.catplot(x='Pclass',kind='count',data = train_df,hue='Person')
```
